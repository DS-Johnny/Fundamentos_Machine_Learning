```python
import cv2
import matplotlib.pyplot as plt
from sklearn import datasets as dt
from sklearn import tree as tr
```

# Load

```python
iris = dt.load_iris()
x = iris.data[:, 2:]
y = iris.target
```

# Model training

```python
# Define

# min_samples_leaf: O número mínimo de amostras do nó deve ter, antes de fazer uma nova separação e gerar nós filhos.
# min_weight_fraction_leaf: A mesma definição do parâmetro
# min_samples_leaf, mas definida como a fração do número total dos pesos das evidências.
# max_features: O número máximo de atributos que são avaliados para a divisão de cada nó.


model_tree = tr.DecisionTreeClassifier(max_depth=10, min_samples_leaf=30)
# Fit
model_tree.fit(x, y)
# Predict
```

```python
tr.export_graphviz(
    model_tree,
    out_file = 'tree.dot',
    feature_names = iris.feature_names[2:],
    class_names = iris.target_names,
    rounded = True,
    filled = True
)
```
