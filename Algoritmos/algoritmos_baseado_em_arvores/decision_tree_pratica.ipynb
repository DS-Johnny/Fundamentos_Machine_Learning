```python
import cv2
import matplotlib.pyplot as plt
from sklearn import datasets as dt
from sklearn import tree as tr
```

# Load

```python
iris = dt.load_iris()
x = iris.data[:, 2:]
y = iris.target
```

# Model training

```python
# Define

# min_samples_leaf: O nÃºmero mÃ­nimo de amostras do nÃ³ deve ter, antes de fazer uma nova separaÃ§Ã£o e gerar nÃ³s filhos.
# min_weight_fraction_leaf: A mesma definiÃ§Ã£o do parÃ¢metro
# min_samples_leaf, mas definida como a fraÃ§Ã£o do nÃºmero total dos pesos das evidÃªncias.
# max_features: O nÃºmero mÃ¡ximo de atributos que sÃ£o avaliados para a divisÃ£o de cada nÃ³.


model_tree = tr.DecisionTreeClassifier(max_depth=10, min_samples_leaf=30)
# Fit
model_tree.fit(x, y)
# Predict
```

```python
tr.export_graphviz(
    model_tree,
    out_file = 'tree.dot',
    feature_names = iris.feature_names[2:],
    class_names = iris.target_names,
    rounded = True,
    filled = True
)
```
