# Imports

```python
from sklearn import datasets as ds
from sklearn import cluster as ct
from sklearn import metrics as mt
import matplotlib.pyplot as plt
import numpy as np
```

# Load Datasets

```python
x, y = ds.make_blobs(n_samples=1000,
              n_features=2,
              centers=5,
              cluster_std=1.2,
              random_state=0)

plt.scatter( x[:,0], x[:,1], c=y)
```

```python
k = 3

# Define
model = ct.KMeans(n_clusters=k, init='random', random_state=0)

# Fit
labels = model.fit_predict(x)

# Silhouette
ss_avg = mt.silhouette_score(x, labels)
ss_avg
```

```python
clusters = np.arange(2, 11, 1)
ss_list = []

for c in clusters:
    # Define
    model = ct.KMeans(n_clusters=c, init='random', random_state=0)

    # Fit
    labels = model.fit_predict(x)

    # Silhouette
    ss_avg = mt.silhouette_score(x, labels)
    ss_list.append(ss_avg)
```

```python
clusters
```

```python
plt.plot(clusters, ss_list, marker='o', linewidth=2)
plt.xlabel("Clusters")
plt.ylabel("Silhouette score")
```
