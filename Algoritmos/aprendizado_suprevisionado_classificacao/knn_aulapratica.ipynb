# SELECT FEATURES

```python
# Features
features = ['idade', 'saldo_atual', 'divida_atual', 'renda_anual', 'valor_em_investimentos', 'taxa_utilizacao_credito', 'num_emprestimos', 'num_contas_bancarias', 'num_cartoes_credito', 'dias_atraso_dt_venc',
     'num_pgtos_atrasados', 'num_consultas_credito', 'taxa_juros']
x_train = df[features].values
# Label
y_train = df.loc[:,'limite_adicional'].values
```

# TRAIN

```python
# DEFINING TRAINMENT PARAMETERS
k = 3
knn_classifier = KNeighborsClassifier( n_neighbors=k )
```

```python
# TRAINING ALGORITHM
knn_classifier.fit( x_train, y_train)
```

```python
df_result = df.copy()
y_pred = knn_classifier.predict( x_train )
df_result['classificacao'] = y_pred
df_result
```

# Rebalancing

```python
from sklearn.utils import resample
df = pd.read_csv('train.csv')
# Separar as classes
class_conceder = df[df['limite_adicional'] == 'Conceder']
class_negar = df[df['limite_adicional'] == 'Negar']
```

```python
# Amostrar aleatoriamente exemplos de cada classe para igualar o número de amostras 50/50
#class_conceder_resampled = resample(class_conceder, replace=True, n_samples=len(class_negar), random_state=42)

# Desbalanceado 90/10
class_conceder_resampled = resample(class_conceder, replace=True, n_samples=int(len(class_negar)*9), random_state=42)
```

```python
# Combinar os exemplos amostrados com os da classe "Negar"
df_balanced = pd.concat([class_conceder_resampled, class_negar])
```

```python
df_balanced.loc[:, 'limite_adicional'].value_counts(normalize=True)
```

```python
# Features
features = ['idade', 'saldo_atual', 'divida_atual', 'renda_anual', 'valor_em_investimentos', 'taxa_utilizacao_credito', 'num_emprestimos', 'num_contas_bancarias', 'num_cartoes_credito', 'dias_atraso_dt_venc',
     'num_pgtos_atrasados', 'num_consultas_credito', 'taxa_juros']
x_train = df_balanced[features].values
# Label
y_train = df_balanced.loc[:,'limite_adicional'].values
```

```python
# DEFINING TRAINMENT PARAMETERS
k = 3
knn_classifier = KNeighborsClassifier( n_neighbors=k )

# TRAINING ALGORITHM
knn_classifier.fit( x_train, y_train)
```

```python
df_result = df_balanced.copy()
y_pred = knn_classifier.predict( x_train )
df_result['classificacao'] = y_pred
df_result
```

```python
# scikit.metrics CONFUSION MATRIX
mt.confusion_matrix(y_train, y_pred)
```

```python

# Accuracy
accuracy = mt.accuracy_score(y_train, y_pred)

# Precision
pres_con = mt.precision_score( y_train, y_pred, average="binary", pos_label='Conceder' )
pres_neg = mt.precision_score( y_train, y_pred, average="binary", pos_label='Negar' )

# Recall
recall_con = mt.recall_score( y_train, y_pred, average='binary', pos_label='Conceder' )
recall_neg = mt.recall_score( y_train, y_pred, average='binary', pos_label='Negar' )

print(30*'-=')
print(f'KNeighbors: {k}')
print(f'Accuracy: {accuracy*100:.2f}%\n')
print(f'Precision:\nConceder: {pres_con*100:.2f}%\nNegar: {pres_neg*100:.2f}%\n')
print(f'Recall:\nConceder: {recall_con*100:.2f}%\nNegar: {recall_neg*100:.2f}%\n')
```
