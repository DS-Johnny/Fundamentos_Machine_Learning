# IMPORT LIBRARIES

```python
# IMPORT LIBRARIES
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier 
from sklearn import metrics as mt
```

# LOAD DATASET

```python
# Load Dataset
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df.loc[:, 'limite_adicional'].value_counts()
```

```python
df.loc[:, 'limite_adicional'].value_counts(normalize=True)
```

# SELECT FEATURES

```python
# Features
features = ['idade', 'saldo_atual', 'divida_atual', 'renda_anual', 'valor_em_investimentos', 'taxa_utilizacao_credito', 'num_emprestimos', 'num_contas_bancarias', 'num_cartoes_credito', 'dias_atraso_dt_venc',
     'num_pgtos_atrasados', 'num_consultas_credito', 'taxa_juros']
x_train = df[features].values
# Label
y_train = df.loc[:,'limite_adicional'].values
```

# TRAIN

```python
# DEFINING TRAINMENT PARAMETERS
k = 3
knn_classifier = KNeighborsClassifier( n_neighbors=k )
```

```python
# TRAINING ALGORITHM
knn_classifier.fit( x_train, y_train)
```

```python
df_result = df.copy()
y_pred = knn_classifier.predict( x_train )
df_result['classificacao'] = y_pred
df_result
```

# PERFORMANCE

```python
# TOTAL ACERTOS VERSUS TOTAL PREVISÕES
right_classification = sum(df_result['limite_adicional'] == df_result['classificacao'])
total_len = len(df_result)
percent_accuracy = right_classification / total_len
print(f'Accuracy: %{percent_accuracy*100:.2f}')
```

```python
conceder = 1505
negar = 7995
acertos_conceder = sum((df_result['limite_adicional'] == 'Conceder') & (df_result['classificacao'] == 'Conceder'))
acertos_negar = sum((df_result['limite_adicional'] == 'Negar') & (df_result['classificacao'] == 'Negar'))

print(f'Conceder Accuracy: {acertos_conceder/conceder*100:.2f}%\nNegar Accuracy: {acertos_negar/negar*100:.2f}%')
```

```python
# scikit.metrics CONFUSION MATRIX
mt.confusion_matrix(y_train, y_pred)
```

## Accuracy

```python
# Accuracy
mt.accuracy_score(y_train, y_pred)
```

## Precision

```python
mt.precision_score( y_train, y_pred, average="binary", pos_label='Conceder' )
```

```python
mt.precision_score( y_train, y_pred, average="binary", pos_label='Negar' )
```

## Recall

```python
mt.recall_score( y_train, y_pred, average='binary', pos_label='Conceder' )
```

```python
mt.recall_score( y_train, y_pred, average='binary', pos_label='Negar' )
```
