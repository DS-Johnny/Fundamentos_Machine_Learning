```python
import matplotlib.pyplot as plt
import numpy as np

# Dados fictícios
x = np.array([1, 2, 3, 4, 5, 9, 12])
y = np.array([2, 4, 5, 4, 5, 13, 9])

# Reta da média
mean_y = np.mean(y)

# Coeficientes da reta de regressão
coefficients = np.polyfit(x, y, 1)
regression_y = np.polyval(coefficients, x)

# Plot
plt.scatter(x, y, color='blue', label='Dados Observados')
plt.plot(x, [mean_y]*len(x), color='red', label='Reta da Média')
plt.plot(x, regression_y, color='green', label='Reta da Regressão')
plt.xlabel('Horas Estudadas')
plt.ylabel('Nota')
plt.legend()
plt.show()
```

Vamos imaginar uma história simples para entender esses conceitos. Imagine que você é um professor que quer analisar a relação entre o número de horas que seus alunos estudam e as notas que eles obtêm em um exame. Você coleta dados de 5 alunos:

Ana estudou 1 hora e tirou nota 2.
Bruno estudou 2 horas e tirou nota 4.
Carla estudou 3 horas e tirou nota 5.
Daniel estudou 4 horas e tirou nota 4.
Elisa estudou 5 horas e tirou nota 5.
Você quer saber quão bem o número de horas estudadas explica as notas dos alunos.

```python
import numpy as np

# Dados de entrada
horas_estudo = np.array([1, 2, 3, 4, 5])
notas = np.array([2, 4, 5, 4, 5])

# Média das notas
media_notas = np.mean(notas)

# Regressão linear
coefficients = np.polyfit(horas_estudo, notas, 1)
notas_preditas = np.polyval(coefficients, horas_estudo)

# Soma dos Quadrados Totais (SSTO)
SSTO = np.sum((notas - media_notas) ** 2)

# Soma dos Quadrados da Regressão (SSR)
SSR = np.sum((notas_preditas - media_notas) ** 2)

# Soma dos Quadrados dos Resíduos (SSE)
SSE = np.sum((notas - notas_preditas) ** 2)

# Coeficiente de Determinação (R²)
R2 = SSR / SSTO

print(f"Média das notas: {media_notas}")
print(f"Coeficientes da regressão (intercepto e inclinação): {coefficients}")
print(f"Notas previstas: {notas_preditas}")
print(f"Soma dos Quadrados Totais (SSTO): {SSTO}")
print(f"Soma dos Quadrados da Regressão (SSR): {SSR}")
print(f"Soma dos Quadrados dos Resíduos (SSE): {SSE}")
print(f"Coeficiente de Determinação (R²): {R2}")
```
